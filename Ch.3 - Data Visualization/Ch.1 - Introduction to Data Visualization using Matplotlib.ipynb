```python
import pandas as pd
import matplotlib.pyplot as plt
```

#### 0 - Data Visualization Example

```python
x = [*range(1,11)]
y = [num**2 for num in x]

plt.plot(x,y)
plt.show()
```

#### 6 - Additional Mandatory

```python
x = [*range(1,11)]
y = [num**2 for num in x]

fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x,y)
plt.show()
```

```python
x = [*range(1,11)]
y = [num**2 for num in x]

fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_title("X and Y relation")
plt.show()
```

```python
x = [*range(1,11)]
y = [num**2 for num in x]

fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_xlabel("Original List")
ax.set_ylabel('y = x ** 2')
plt.show()
```
