```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.simplefilter('ignore')
```

```python
url = "https://raw.githubusercontent.com/SophrosyneEunoia/Shockwave/main/sector.txt"

sector = pd.read_csv(url)
sector.head()
```

<p style="font-size:20px">EPS<br>Earning Per Share<br>Suatu pendapatan bersih milik perusahaan dalam satu tahun yang dibagi dengan jumlah rata-rata lembaran saham yang sudah diedarkan</p>

```python
sector.info()
```

<p style="font-size:20px">Top 5 Highest Price for Information Technology Sector</p>

```python
it_sector = sector[sector['Sector']=='Information Technology']
it_sector.sort_values(by='Price', ascending=False, inplace=True)
it_sector.head()
```

<p style="font-size:20px">Top 5 Highest Price</p>

```python
sector.sort_values(by='Price', ascending=False, inplace=True)
sector.head()
```

<p style="font-size:20px">Top 3 Highest Price Sector</p>

```python
groupby_sector = sector.groupby('Sector').mean()
groupby_sector.sort_values(by='Price', ascending=False, inplace=True)
groupby_sector.head()
```

<p style="font-size:20px">P/E Ratio</p>

```python
sector.head(3)
```

```python
sector['P/E Ratio'] = round(sector['Price'] / sector['EPS'],2)
sector.head()
```

<p style="font-size:20px">Top 3 Highest P/E Ratio Sector</p>

```python
groupby_sector = sector.groupby('Sector').mean()
groupby_sector.sort_values(by='P/E Ratio', ascending=False, inplace=True)
groupby_sector.head()
```

<p style="font-size:20px">5 Highest P/E Ratio Company</p>

```python
sector_pe = sector.sort_values(by='P/E Ratio',ascending=False)
sector_pe.head()
```
