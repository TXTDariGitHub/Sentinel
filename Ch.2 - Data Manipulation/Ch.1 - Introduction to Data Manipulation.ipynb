### 2 - Aggregating DataFrames

```python
url_sales = "https://raw.githubusercontent.com/UvoChrollo/Shockwave/main/sales_subset.csv"

sales = pd.read_csv(url_sales)
sales.head()
```

```python
sales['weekly_sales'].describe()
```

```python
sales['weekly_sales'].mean()
```

```python
sales['weekly_sales'].agg(['mean','median'])
```

```python
sales[['weekly_sales']].agg(['mean','median'])
```

```python
sales[['weekly_sales','unemployment']].agg(['max','min'])
```

```python
def iqr(column):
    return column.quantile(0.75) - column.quantile(0.25)
    
print(sales["weekly_sales"].agg(iqr))
```

```python
sales[['weekly_sales','unemployment']].agg(iqr)
```

```python
sales.sort_values('weekly_sales',inplace=True)
sales.head()
```

```python
sales['cum_weekly'] = sales['weekly_sales'].cumsum()
sales.head()
```

```python
s_type = sales.drop_duplicates(subset='type')
s_type.head()
```

```python
s_str_type = sales.drop_duplicates(subset=['store','type'])
s_str_type
```

### 4 - Creating and Visualizing DataFrames

```python
import pandas as pd
import matplotlib.pyplot as plt

avocado = pd.read_pickle("https://github.com/UvoChrollo/Shockwave/blob/main/avoplotto.pkl?raw=true")
avocado.head()
```

```python
avocado['date'] = pd.to_datetime(avocado['date'])
avocado.head()
```

```python
avocado['date'].dtype
```

```python
avocado.sort_values(by='date',inplace=True)
avocado.head()
```

```python
nb_sold_by_size = avocado.groupby("size")["nb_sold"].sum()
nb_sold_by_size.plot(kind="bar")
plt.show()
```

```python
avocado.groupby('date')['nb_sold'].sum().plot(kind='line')
plt.show()
```

```python
avocado.groupby('date')['nb_sold'].sum().plot(kind='line')
plt.title("Penjualan Perhari")
plt.show()
```

```python
avocado.groupby('year')['nb_sold'].sum().plot(kind='line')
plt.title("Penjualan Pertahun")
plt.show()
```

```python
avocado.plot(x="nb_sold", y="avg_price", kind="scatter", title="Number of avocados sold vs. average price")
plt.show()
```

```python
avocado[avocado["type"] == "conventional"]["avg_price"].hist()
avocado[avocado["type"] == "organic"]["avg_price"].hist()
plt.legend(["conventional", "organic"])
plt.show()
```
